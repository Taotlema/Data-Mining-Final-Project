```python
df_data = pd.get_dummies(df_data, columns=['EXPORT_CNTRY'], prefix='EXPORT_CNTRY', drop_first=False, dtype=int)
display(df_data.head())
df_data.info()
```

```python
df_data = pd.get_dummies(df_data, columns=['HAZMAT'], prefix='HAZMAT', drop_first=False, dtype=int)
display(df_data.head())
df_data.info()
```

```python
print("Value counts for mode_series after filtering:")
print(df_data['MODE'].value_counts())
```

```python
filtered_indices = df_data[df_data['MODE'].isin([4, 5, 14])].index
df_data = df_data.loc[filtered_indices]

print(f"Number of rows after filtering: {df_data.shape[0]}")
print(f"Number of columns after filtering: {df_data.shape[1]}")
```

```python
print("Value counts for mode_series after filtering:")
print(df_data['MODE'].value_counts())
```

```python
mode_series = df_data['MODE']
display(mode_series.head())
```

```python
df_data = df_data.drop(columns=['MODE'])
df_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Define features (X) and target (y)
X = df_data
y = mode_series

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10000, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
# Split off another 10,000 examples for a development set from the training data
X_train_dev, X_dev, y_train_dev, y_dev = train_test_split(X_train, y_train, test_size=10000, random_state=42, stratify=y_train)

print(f"Shape of new X_train (dev): {X_train_dev.shape}")
print(f"Shape of X_dev: {X_dev.shape}")
print(f"Shape of new y_train (dev): {y_train_dev.shape}")
print(f"Shape of y_dev: {y_dev.shape}")
```

```python
X_train.isna().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Impute missing values in WGT_FACTOR before training
X_train_dev['WGT_FACTOR'].fillna(X_train_dev['WGT_FACTOR'].mean(), inplace=True)
X_dev['WGT_FACTOR'].fillna(X_dev['WGT_FACTOR'].mean(), inplace=True)
X_test['WGT_FACTOR'].fillna(X_test['WGT_FACTOR'].mean(), inplace=True)

# --- Logistic Regression Model ---
print("\n--- Logistic Regression Model ---")
# Initialize the Logistic Regression model
log_reg_model = LogisticRegression(max_iter=10000, random_state=42, solver='liblinear')

# Train the model
log_reg_model.fit(X_train_dev, y_train_dev)

# Make predictions on the development set
y_pred_log_reg = log_reg_model.predict(X_dev)

# Evaluate the Logistic Regression model
print(f"Logistic Regression Accuracy: {accuracy_score(y_dev, y_pred_log_reg):.4f}")
print("Logistic Regression Classification Report:")
print(classification_report(y_dev, y_pred_log_reg))
```

```python
# --- Decision Tree Model ---
print("\n--- Decision Tree Model ---")
# Initialize the Decision Tree Classifier
dec_tree_model = DecisionTreeClassifier(random_state=42)

# Train the model
dec_tree_model.fit(X_train_dev, y_train_dev)

# Make predictions on the development set
y_pred_dec_tree = dec_tree_model.predict(X_dev)

# Evaluate the Decision Tree model
print(f"Decision Tree Accuracy: {accuracy_score(y_dev, y_pred_dec_tree):.4f}")
print("Decision Tree Classification Report:")
print(classification_report(y_dev, y_pred_dec_tree))
```
